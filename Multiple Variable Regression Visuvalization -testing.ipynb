```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

df_adv = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
df_adv.head()
```

```python
X = df_adv[['TV', 'radio']]
y = df_adv['sales']
df_adv.head()
```

```python
X = sm.add_constant(X)
est = sm.OLS(y, X).fit()
est.summary()
```
